Waist Circumference vs. BMI: The article strongly suggests that waist circumference is a better predictor of health outcomes (including risk of death and cardiovascular disease) than BMI, especially when both are considered together.

Combined Use: The combination of BMI and waist circumference is recommended to identify a "high-risk obesity phenotype" better than either measure alone.


In summary, the articles emphasize that while BMI is a general measure of obesity, waist circumference inclusion provides crucial additional information about abdominal obesity and its associated health risks, even within normal BMI ranges. #####  The data processing steps:

Select relevant columns: From both df_bmx_c and df_bmx_d, we will select SEQN, BMXWT, BMXHT, BMXWAIST, and BMXBMI.


Add survey_year column: Add a survey_year column to each DataFrame ('2003-2004' for df_bmx_c and '2005-2006' for df_bmx_d).

Calculate BMI and Waist-to-Height Ratio:
BMI: We will calculate BMI_calculated = BMXWT / (BMXHT / 100)**2. We can then compare this with BMXBMI to decide whether to use the pre-calculated one or our own. Usually, using the pre-calculated one is fine, but it's good to verify. For now, let's just create BMI_calculated.


Waist-to-Height Ratio (WHtR): Calculate WHtR = BMXWAIST / BMXHT.

```python
import pandas as pd

# --- File Paths for BMX Data ---
# Direct download link for BMX_C.csv 
bmx_c_google_drive_link = 'https://drive.google.com/file/d/1L4ELPxCdAhrCOdGCuRUeUqsDR1WfwN76/view?usp=sharing'
bmx_c_file_id = bmx_c_google_drive_link.split('/')[-2]
bmx_c_direct_download_link = f'https://drive.google.com/uc?export=download&id={bmx_c_file_id}'

# Direct download linkfor BMX_D.csv (from initial query)
bmx_d_google_drive_link = 'https://drive.google.com/file/d/1euU6BHI2ncrHNzeos8_3njcw_8VGLJwQ/view?usp=sharing'
bmx_d_file_id = bmx_d_google_drive_link.split('/')[-2]
bmx_d_direct_download_link = f'https://drive.google.com/uc?export=download&id={bmx_d_file_id}'


try:
    # Load the BMX_C data
    df_bmx_c = pd.read_csv(bmx_c_direct_download_link)
    # Load the BMX_D data
    df_bmx_d = pd.read_csv(bmx_d_direct_download_link)

    # 1. Select relevant columns and 2. Add 'survey_year' column
    selected_cols = ['SEQN', 'BMXWT', 'BMXHT', 'BMXWAIST', 'BMXBMI']

    df_bmx_c_processed = df_bmx_c[selected_cols].copy()
    df_bmx_c_processed['survey_year'] = '2003-2004'

    df_bmx_d_processed = df_bmx_d[selected_cols].copy()
    df_bmx_d_processed['survey_year'] = '2005-2006'

    # 3. Merge the DataFrames
    merged_bmx_data = pd.concat([df_bmx_c_processed, df_bmx_d_processed], ignore_index=True)

    print("--- Merged BMX Data (merged_bmx_data) Head ---")
    print(merged_bmx_data.head())
    print("\n--- Merged BMX Data Info ---")
    print(merged_bmx_data.info())

    # 4. Calculate BMI and Waist-to-Height Ratio
    # Convert height from cm to meters for BMI calculation
    merged_bmx_data['BMXHT_m'] = merged_bmx_data['BMXHT'] / 100

    # Calculate BMI
    # BMI = weight (kg) / height (m)^2
    merged_bmx_data['BMI_calculated'] = merged_bmx_data['BMXWT'] / (merged_bmx_data['BMXHT_m']**2)

    # Calculate Waist-to-Height Ratio (WHtR)
    # WHtR = Waist Circumference (cm) / Height (cm)
    merged_bmx_data['WHtR_calculated'] = merged_bmx_data['BMXWAIST'] / merged_bmx_data['BMXHT']

    print("\n--- Merged BMX Data with Calculated Metrics Head ---")
    print(merged_bmx_data.head())

    print("\n--- Descriptive Statistics for Calculated BMI and WHtR ---")
    print("BMI_calculated:\n", merged_bmx_data['BMI_calculated'].describe())
    print("\nWHtR_calculated:\n", merged_bmx_data['WHtR_calculated'].describe())

    # Compare pre-calculated BMXBMI with BMI_calculated
    # This comparison helps decide if we should use the pre-calculated one or our own
    print("\n--- Comparison of BMXBMI (NHANES) vs. BMI_calculated (Our own) ---")
    # For a subset of non-null values for comparison
    comparison_df = merged_bmx_data[['BMXBMI', 'BMI_calculated']].dropna()
    if not comparison_df.empty:
        # Calculate the absolute difference
        comparison_df['BMI_diff'] = abs(comparison_df['BMXBMI'] - comparison_df['BMI_calculated'])
        print(f"Mean absolute difference between NHANES BMI and calculated BMI: {comparison_df['BMI_diff'].mean():.4f}")
        print(f"Max absolute difference: {comparison_df['BMI_diff'].max():.4f}")
        print(f"Number of cases where BMI_diff > 0.01: {len(comparison_df[comparison_df['BMI_diff'] > 0.01])}")
    else:
        print("Not enough non-null data to compare BMXBMI and BMI_calculated.")


except Exception as e:
    print(f"An error occurred during BMX data processing: {e}")
    
```

##### Impute missing values for BMXWT, BMXHT, BMXWAIST, and BMXBMI using their respective medians. After these are imputed, we will then recalculate WHtR_calculated using the newly imputed BMXHT and BMXWAIST to ensure consistency.

```python
import pandas as pd
import numpy as np

# --- Re-create merged_bmx_data with calculations 
# Direct download link for BMX_C.csv
bmx_c_google_drive_link = 'https://drive.google.com/file/d/1L4ELPxCdAhrCOdGCuRUeUqsDR1WfwN76/view?usp=sharing'
bmx_c_file_id = bmx_c_google_drive_link.split('/')[-2]
bmx_c_direct_download_link = f'https://drive.google.com/uc?export=download&id={bmx_c_file_id}'

# Direct download link for BMX_D.csv
bmx_d_google_drive_link = 'https://drive.google.com/file/d/1euU6BHI2ncrHNzeos8_3njcw_8VGLJwQ/view?usp=sharing'
bmx_d_file_id = bmx_d_google_drive_link.split('/')[-2]
bmx_d_direct_download_link = f'https://drive.google.com/uc?export=download&id={bmx_d_file_id}'

try:
    df_bmx_c = pd.read_csv(bmx_c_direct_download_link)
    df_bmx_d = pd.read_csv(bmx_d_direct_download_link)

    selected_cols = ['SEQN', 'BMXWT', 'BMXHT', 'BMXWAIST', 'BMXBMI']
    df_bmx_c_processed = df_bmx_c[selected_cols].copy()
    df_bmx_c_processed['survey_year'] = '2003-2004'
    df_bmx_d_processed = df_bmx_d[selected_cols].copy()
    df_bmx_d_processed['survey_year'] = '2005-2006'
    merged_bmx_data = pd.concat([df_bmx_c_processed, df_bmx_d_processed], ignore_index=True)

    # Initial calculation (will be updated after imputation of base measures)
    merged_bmx_data['BMXHT_m'] = merged_bmx_data['BMXHT'] / 100
    merged_bmx_data['BMI_calculated'] = merged_bmx_data['BMXWT'] / (merged_bmx_data['BMXHT_m']**2)
    merged_bmx_data['WHtR_calculated'] = merged_bmx_data['BMXWAIST'] / merged_bmx_data['BMXHT']

    print("--- Performing Median Imputation on BMX Variables ---")

    # Impute BMXWT, BMXHT, BMXWAIST, and BMXBMI with their respective medians
    # Use .fillna() and assign back to the column to avoid SettingWithCopyWarning
    merged_bmx_data['BMXWT'] = merged_bmx_data['BMXWT'].fillna(merged_bmx_data['BMXWT'].median())
    merged_bmx_data['BMXHT'] = merged_bmx_data['BMXHT'].fillna(merged_bmx_data['BMXHT'].median())
    merged_bmx_data['BMXWAIST'] = merged_bmx_data['BMXWAIST'].fillna(merged_bmx_data['BMXWAIST'].median())
    merged_bmx_data['BMXBMI'] = merged_bmx_data['BMXBMI'].fillna(merged_bmx_data['BMXBMI'].median())

    # Recalculate WHtR_calculated after BMXHT and BMXWAIST have been imputed
    merged_bmx_data['BMXHT_m'] = merged_bmx_data['BMXHT'] / 100 # Recalculate height in meters as well
    merged_bmx_data['WHtR_calculated'] = merged_bmx_data['BMXWAIST'] / merged_bmx_data['BMXHT']

    # We will use BMXBMI directly and not BMI_calculated as agreed earlier.
    # So we don't need to recalculate BMI_calculated here, just ensuring BMXBMI is imputed.


    # Verify no more missing values for the imputed columns
    imputed_cols = ['BMXWT', 'BMXHT', 'BMXWAIST', 'BMXBMI', 'WHtR_calculated'] # Check WHtR too as it was recalculated
    missing_after_imputation = merged_bmx_data[imputed_cols].isnull().sum()
    print("\nMissing values after imputation:")
    print(missing_after_imputation)

    print("\n--- Descriptive Statistics for Key BMX Variables AFTER Imputation ---")
    for col in imputed_cols:
        print(f"\n--- {col} ---")
        print(merged_bmx_data[col].describe())
        # Print info to see non-null count matches total rows
        # print(f"Non-null count for {col}: {merged_bmx_data[col].count()}")


except Exception as e:
    print(f"An error occurred during BMX data imputation: {e}")
  
```

```python
import pandas as pd
import numpy as np

# --- Re-load and prepare merged_crp_data (with imputation and transformation) ---
crp_google_drive_link = 'https://drive.google.com/file/d/1qFASkuMJRjVqsOJC2FvElL5IlL1tdeAK/view?usp=sharing'
crp_file_id = crp_google_drive_link.split('/')[-2]
crp_direct_download_link = f'https://drive.google.com/uc?export=download&id={crp_file_id}'

merged_crp_data = pd.read_csv(crp_direct_download_link)
crp_median = merged_crp_data['LBXCRP'].median()
merged_crp_data['LBXCRP'] = merged_crp_data['LBXCRP'].fillna(crp_median)
merged_crp_data['log_LBXCRP'] = np.log1p(merged_crp_data['LBXCRP'])

# --- Re-load and prepare merged_bmx_data (with imputation and calculated WHtR) ---
bmx_c_google_drive_link = 'https://drive.google.com/file/d/1L4ELPxCdAhrCOdGCuRUeUqsDR1WfwN76/view?usp=sharing'
bmx_c_file_id = bmx_c_google_drive_link.split('/')[-2]
bmx_c_direct_download_link = f'https://drive.google.com/uc?export=download&id={bmx_c_file_id}'

bmx_d_google_drive_link = 'https://drive.google.com/file/d/1euU6BHI2ncrHNzeos8_3njcw_8VGLJwQ/view?usp=sharing'
bmx_d_file_id = bmx_d_google_drive_link.split('/')[-2]
bmx_d_direct_download_link = f'https://drive.google.com/uc?export=download&id={bmx_d_file_id}'

df_bmx_c = pd.read_csv(bmx_c_direct_download_link)
df_bmx_d = pd.read_csv(bmx_d_direct_download_link)

selected_bmx_cols = ['SEQN', 'BMXWT', 'BMXHT', 'BMXWAIST', 'BMXBMI']
df_bmx_c_processed = df_bmx_c[selected_bmx_cols].copy()
df_bmx_c_processed['survey_year'] = '2003-2004'
df_bmx_d_processed = df_bmx_d[selected_bmx_cols].copy()
df_bmx_d_processed['survey_year'] = '2005-2006'
merged_bmx_data = pd.concat([df_bmx_c_processed, df_bmx_d_processed], ignore_index=True)

# Perform Median Imputation on BMX Variables (before final merge)
merged_bmx_data['BMXWT'] = merged_bmx_data['BMXWT'].fillna(merged_bmx_data['BMXWT'].median())
merged_bmx_data['BMXHT'] = merged_bmx_data['BMXHT'].fillna(merged_bmx_data['BMXHT'].median())
merged_bmx_data['BMXWAIST'] = merged_bmx_data['BMXWAIST'].fillna(merged_bmx_data['BMXWAIST'].median())
merged_bmx_data['BMXBMI'] = merged_bmx_data['BMXBMI'].fillna(merged_bmx_data['BMXBMI'].median())

# Recalculate WHtR_calculated
merged_bmx_data['WHtR_calculated'] = merged_bmx_data['BMXWAIST'] / merged_bmx_data['BMXHT']


# --- Load Demographic Files ---
demo_c_google_drive_link = 'https://drive.google.com/file/d/1U6iSDa9F0mkXhwBlKA1eDv8TEJlbSTOX/view?usp=sharing'
demo_c_file_id = demo_c_google_drive_link.split('/')[-2]
demo_c_direct_download_link = f'https://drive.google.com/uc?export=download&id={demo_c_file_id}'

demo_d_google_drive_link = 'https://drive.google.com/file/d/17VtHsL7vA8pNTA5Anltzzq3SMWNGJa1z/view?usp=sharing'
demo_d_file_id = demo_d_google_drive_link.split('/')[-2]
demo_d_direct_download_link = f'https://drive.google.com/uc?export=download&id={demo_d_file_id}'

try:
    df_demo_c = pd.read_csv(demo_c_direct_download_link)
    df_demo_d = pd.read_csv(demo_d_direct_download_link)

    selected_demo_cols = ['SEQN', 'RIDAGEYR', 'RIAGENDR', 'RIDEXMON', 'RIDAGEMN', 'RIDAGEEX']

    df_demo_c_processed = df_demo_c[selected_demo_cols].copy()
    df_demo_c_processed['survey_year'] = '2003-2004'

    df_demo_d_processed = df_demo_d[selected_demo_cols].copy()
    df_demo_d_processed['survey_year'] = '2005-2006'

    merged_demo_data = pd.concat([df_demo_c_processed, df_demo_d_processed], ignore_index=True)

    merged_demo_data['RIAGENDR'] = merged_demo_data['RIAGENDR'].map({1: 'Male', 2: 'Female'})

    # --- Merge All DataFrames ---
    merged_crp_data['SEQN'] = merged_crp_data['SEQN'].astype(float)
    merged_bmx_data['SEQN'] = merged_bmx_data['SEQN'].astype(float)
    merged_demo_data['SEQN'] = merged_demo_data['SEQN'].astype(float)

    combined_data = pd.merge(merged_crp_data, merged_bmx_data, on=['SEQN', 'survey_year'], how='outer')
    merged_demo_data_for_merge = merged_demo_data.rename(columns={'survey_year': 'demo_survey_year'}).copy()
    final_merged_data = pd.merge(combined_data, merged_demo_data_for_merge, on='SEQN', how='outer')

    if 'demo_survey_year' in final_merged_data.columns:
        final_merged_data['survey_year'] = final_merged_data['survey_year'].fillna(final_merged_data['demo_survey_year'])
        final_merged_data.drop(columns=['demo_survey_year'], inplace=True)

    # --- Reorder columns ---
    new_column_order = [
        'SEQN', 'survey_year', 'RIDAGEYR', 'RIAGENDR', 'RIDEXMON', 'RIDAGEMN', 'RIDAGEEX',
        'BMXWT', 'BMXHT', 'BMXWAIST', 'BMXBMI', 'WHtR_calculated',
        'LBXCRP', 'log_LBXCRP'
    ]
    new_column_order_filtered = [col for col in new_column_order if col in final_merged_data.columns]
    remaining_columns = [col for col in final_merged_data.columns if col not in new_column_order_filtered]
    final_column_order = new_column_order_filtered + sorted(remaining_columns)
    final_merged_data = final_merged_data[final_column_order]

    # --- Perform Final Imputation on All Remaining Missing Numerical Columns ---
    print("--- Performing Final Median Imputation on All Remaining Missing Numerical Columns ---")
    cols_to_impute_final = [
        'LBXCRP', 'log_LBXCRP', 'RIDAGEEX', 'RIDEXMON',
        'BMXWT', 'BMXHT', 'BMXWAIST', 'BMXBMI', 'WHtR_calculated', 'RIDAGEMN'
    ]

    for col in cols_to_impute_final:
        if col in final_merged_data.columns and final_merged_data[col].isnull().any():
            median_val = final_merged_data[col].median()
            final_merged_data[col] = final_merged_data[col].fillna(median_val)
            print(f"Imputed '{col}' with median value: {median_val:.2f}")
        elif col in final_merged_data.columns:
            print(f"'{col}' has no missing values, skipping imputation.")
        else:
            print(f"Warning: Column '{col}' not found in DataFrame for imputation.")

    # Create the new DataFrame copy
    imputed_final_data = final_merged_data.copy()

    # Verify that there are no more missing values in the new DataFrame
    print("\n--- Missing Values in 'imputed_final_data' after final imputation ---")
    print(imputed_final_data.isnull().sum().sort_values(ascending=False))

    # --- Save to CSV ---
    output_file_name = '../data/processed_data/nhanes_final_imputed_data.csv'
    
    imputed_final_data.to_csv(output_file_name, index=False)

    print(f"\n--- Final, fully imputed data saved to '{output_file_name}' ---")
   


except Exception as e:
    print(f"An error occurred during final imputation and saving: {e}")
  
    
```
